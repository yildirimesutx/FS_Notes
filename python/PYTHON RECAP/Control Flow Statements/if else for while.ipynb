```python
# girilen bir cümlenin en uzun karakterli olanı bulun

cümle = input("Bir cümle giriniz : ")

cümle_list = cümle.split()

i = 0

longest = 0

while i < len(cümle_list) :
    
    if len(cümle_list[i]) > longest :    # bu şartı sağladıktan sonra longest yeni değerini alıyor,  
                                         # 10 karakterli olan girildi longest 10 oluyor ve girilen değerde 10 şartı sağlanan kadar en büyük o kalıyor
        longest = len(cümle_list[i])
        
    
    i += 1   
    
print("En uzun karakterli kelime : ", longest , "harflidir")   
```

```python
# Task : Write a program that takes a number from the user and prints the result to check if it is a prime number.

n = int(input("enter a number: "))

counter = 0

for i in range(1, n+1):
   if n % i == 0 :
      counter += 1

if (n==0) or (n==1) or (counter >= 3) :
    print(n, "is not a Prime number")

else :
  print(n, "is a Prime number")
```
